```python
def plot_annot_mrc(mrc_file_path, particle_annotation):
    mrc_data = mrcfile.read(mrc_file_path)
    annot = pd.read_csv(particle_annotation)
    
    plt.imshow(mrc_data, cmap='gray')
    plt.grid(False)
    annot['Y_fixed'] = mrc_data.shape[0] - annot['Y-Coordinate'] - 1  # Adjust Y-coordinates for image display
    plt.scatter(annot['X-Coordinate'], annot['Y_fixed'], c='green', s=10)
    plt.show()
```

```python
plot_annot_mrc(mrc_file_path_template.format('0002'), particle_annotation_template.format('0002'))
```
